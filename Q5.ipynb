# Question

Build a model to classify images of blood cells into 8 categories.

The model should input images size 60x60.

Normalizing the data will help you train a more powerful model.

Please save the model in the location given by **model_path**.

```python
!wget https://utils-computervisiondeveloper.s3.amazonaws.com/media/public/TensorFlowDeveloperCertificate/submit_and_test_model_q5.py url -O /tmp/submit_and_test_model_q5.py

import sys
sys.path.append('/tmp/')
```

```python
model_path = '/tmp/model.h5'
```

```python
# We will be using bloodmnist!
# You can find the dataset here: https://medmnist.com/
# Jiancheng Yang, Rui Shi, Donglai Wei, Zequan Liu, Lin Zhao, Bilian Ke, Hanspeter Pfister, Bingbing Ni. Yang, Jiancheng, et al.
# MedMNIST v2-A large-scale lightweight benchmark for 2D and 3D biomedical image classification." Scientific Data, 2023.

!wget https://utils-computervisiondeveloper.s3.amazonaws.com/media/public/TensorFlowDeveloperCertificate/bloodmnist-dataset.zip

!unzip /content/bloodmnist-dataset.zip -d /tmp/
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1.0/255
)
target_size = (60, 60)
class_mode = 'categorical'

train_data_generator = datagen.flow_from_directory(
    '/tmp/bloodmnist-dataset/train',
    target_size= target_size,
    batch_size=32,
    class_mode= class_mode
)

validation_data_generator = datagen.flow_from_directory(
    '/tmp/bloodmnist-dataset/val',
    target_size= target_size,
    batch_size=32,
    class_mode= class_mode
)
```

```python
from tensorflow.keras import Sequential


model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(60, 60, 3)),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(8, activation='softmax')  # Assuming BloodMNIST has 8 classes
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(
    train_data_generator,
    epochs=10,
    validation_data=validation_data_generator
)

model.save(model_path)
```

```python
# Save the model
model_path = '/tmp/bloodmnist_cnn_model.h5'
model.save(model_path)
```
