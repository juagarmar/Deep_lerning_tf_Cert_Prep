# Question

Build a model to classify images of ships into 2 categories.

The model should input images of size 80x80.

Normalizing the data will help you train a more powerful model.

Please save the model in the location given by **model_path**.

```python
model_path = '/tmp/model.h5'
```

```python
# Satellite imagery used to build this dataset is made available through Planet's Open California dataset, which is openly licensed.
# As such, this dataset is also available under the same CC-BY-SA license. Users can sign up for a free Planet account to search, view,
# and download their imagery and gain access to their API.
# Additionally, the dataset was downloaded from this post: https://www.kaggle.com/datasets/rhammell/ships-in-satellite-imagery/data
# The data was arranged into three sets (train, val and test). Only a subset of images was considered.

!wget https://utils-computervisiondeveloper.s3.amazonaws.com/media/public/TensorFlowDeveloperCertificate/ship-image-classification-data.zip

!unzip /content/ship-image-classification-data.zip -d /tmp/
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values to [0, 1]
)

train_data_generator = datagen.flow_from_directory(
    '/tmp/ship-image-classification-data/train',
    target_size=(80, 80),
    batch_size=32,
    class_mode='binary'
)

validation_data_generator = datagen.flow_from_directory(
    '/tmp/ship-image-classification-data/val',
    target_size=(80, 80),
    batch_size=32,
    class_mode='binary'
)
```

```python
from tensorflow.keras import Sequential
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define the CNN model
model = Sequential([
    Conv2D(32, kernel_size=(5, 5), activation='relu', input_shape=(80, 80, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(5, 5), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # Single output neuron with sigmoid activation for binary classification
])

model.save(model_path)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_data_generator, epochs=10, validation_data=validation_data_generator)
```

```python
# Save the model
model_path = 'cnn_model.h5'
model.save(model_path)
```
